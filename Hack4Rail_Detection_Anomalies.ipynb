Writing a function, that classifies the day as usual and unsusual

```python
def classify_day(pax):
    if pax < lower_bound or pax > upper_bound:
        return 'Unusual'
    else:
        return 'Usual'

all_the_days = []

for weekday in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']:
    weekdays_df = ffmO_df[ffmO_df['weekday'] == weekday].copy()

    # Sum data_pax per day
    daily_totals = weekdays_df.groupby('date')['data_pax'].sum().reset_index(name='total_pax')

    mean_total = daily_totals['total_pax'].mean()
    std_total = daily_totals['total_pax'].std()
    lower_bound = mean_total - std_total
    upper_bound = mean_total + std_total

    # Classify each day
    daily_totals['status'] = daily_totals['total_pax'].apply(classify_day)
    daily_totals['date'] = pd.to_datetime(daily_totals['date']).dt.date
    weekdays_df['date'] = pd.to_datetime(weekdays_df['date']).dt.date

    # Create a dictionary date -> {status, total_pax, mean, std}
    daily_info_dict = {}
    for _, row in daily_totals.iterrows():
        date_str = row['date'].strftime('%Y-%m-%d')
        daily_info_dict[date_str] = {
            'status': row['status'],
            'total_pax': row['total_pax'],
            'mean': mean_total,
            'std': std_total
        }
    print(daily_info_dict)
    # Assign values to weekdays_df
    weekdays_df['date_str'] = weekdays_df['date'].apply(lambda d: d.strftime('%Y-%m-%d'))

    weekdays_df['status'] = weekdays_df['date_str'].map(lambda d: daily_info_dict.get(d, {}).get('status'))
    weekdays_df['weekday_mean'] = weekdays_df['date_str'].map(lambda d: daily_info_dict.get(d, {}).get('mean'))
    weekdays_df['weekday_std'] = weekdays_df['date_str'].map(lambda d: daily_info_dict.get(d, {}).get('std'))
    weekdays_df['total_pax_day'] = weekdays_df['date_str'].map(lambda d: daily_info_dict.get(d, {}).get('total_pax'))
    weekdays_df['weekday'] = weekday

    # Drop the helper column
    weekdays_df.drop(columns=['date_str'], inplace=True)

    all_the_days.append(weekdays_df)

# Combine all weekdays into one dataframe
df_processed = pd.concat(all_the_days)

# Display result
print(df_processed)
print(df_processed['status'].value_counts())
```

```python
df_processed.to_csv("ffmO_processed.csv", index=False)
```
